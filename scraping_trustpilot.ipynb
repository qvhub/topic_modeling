```python
%time

import json
import pandas as pd
import requests
from bs4 import BeautifulSoup
import random
from tqdm.auto import tqdm
import time
import os
from datetime import datetime
import numpy as np
import sys
import re 

#import re

#my_url = "https://fr.trustpilot.com//review/www.e-leclerc.com"


ls_usr_name = []
ls_usr_nbr_reviews = []
ls_usr_locality = []
ls_stars = []
ls_review_date_posted = []
ls_review_title = []
ls_review = []
ls_review_date_experience = []
page = 1
url = "https://fr.trustpilot.com/review/www.ldlc.com"#https://fr.trustpilot.com/review/totalenergies.fr"#"https://fr.trustpilot.com/review/www.amazon.fr"
last_page = ''


while True:
    #print(page)

    user_agent = {'User-agent': 'Mozilla/5.0'}

    # appel de la fonction pour executer le JS
    results = requests.get(url, headers = user_agent)

    #soup = BeautifulSoup(results.text.replace('script', 'myscript'), "lxml")
    soup = BeautifulSoup(results.text, "html.parser")

    if last_page == '':
        last_page = int(soup.find(attrs={"data-pagination-button-last-link":"true"}).text)

    ls_article = soup.find('section', class_="styles_reviewsContainer__3_GQw").find_all("article")
    

    for article in ls_article:

        usr_name = article.aside.a.div.text
        usr_nbr_reviews = article.aside.a.span.text
        usr_locality = article.aside.find_all(attrs={"data-consumer-country-typography":"true"})[0].text

        stars = ls_article[0].find('div', class_="star-rating_starRating__4rrcf").img['alt']
        stars = re.findall(r'\b\d+\b', stars)[0]

        review_date_posted = article.find('time')['datetime']

        if article.find(attrs={"data-review-title-typography":"true"}) == None:
            review_title = np.nan
        else:
            review_title = article.find(attrs={"data-review-title-typography":"true"}).text

        if article.find(attrs={"data-service-review-text-typography":"true"}) == None:
            review = np.nan
        else:
            review = article.find(attrs={"data-service-review-text-typography":"true"}).text

        if article.find(attrs={"data-service-review-date-of-experience-typography":"true"}) == None:
            review_date_experience = np.nan
        else:
            review_date_experience = article.find(attrs={"data-service-review-date-of-experience-typography":"true"}).text.split(':')[1].strip()

        ls_usr_name.append(usr_name)
        ls_usr_nbr_reviews.append(usr_nbr_reviews)
        ls_usr_locality.append(usr_locality)
        ls_stars.append(stars)
        ls_review_date_posted.append(review_date_posted)
        ls_review_title.append(review_title)
        ls_review.append(review)
        ls_review_date_experience.append(review_date_experience)

    if page == last_page:
        break
    else:
        page += 1
        url = "https://fr.trustpilot.com/review/www.ldlc.com?page=" + str(page)
    
    time.sleep(random.uniform(0.1, 2.5))

df = pd.DataFrame({
'usr_name': ls_usr_name,
'usr_nbr_reviews': ls_usr_nbr_reviews,
'usr_locality': ls_usr_locality,
'stars': ls_stars,
'review_date_posted': ls_review_date_posted,
'review_title': ls_review_title,
'review': ls_review,
'review_date_experience': ls_review_date_experience
})

file_name = url.split('review/')[1].split('/page=')[0].replace('.', '_').replace('?', '_') + '.tsv'
df.to_csv(file_name, sep='\t', na_rep='NULL')
```
